```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8s.pt')
```

```python
hyperparameters={
    'data': '/content/drive/MyDrive/Colab Notebooks/dataset2/data.yaml',
    'epochs': 100,
    'batch': 16,
    'imgsz': 640,
    'patience': 20,
    'optimizer': 'Adam',
    'lr0': 0.001,
    'lrf': 0.01,
    'cos_lr': True,
    'weight_decay': 5e-4,
    'box': 7.5,
    'cls': 0.5,
    'dfl': 1.5,

    'save': True,
    'save_period': 5,
    'project': "/content/drive/MyDrive/weapon_detection",
    'name': 'weapon_detector',
    'exist_ok': True,

    'device': 0,
    'workers': 8,
    'verbose': True,
    'amp': True,
    'cache': None
}
```

```python
model.train(**hyperparameters)
```

```python
model.val()
```

```python
results = model.predict(
    source="/content/drive/MyDrive/Colab Notebooks/dataset2/test/images",
    conf=0.3,
    save=True,
    save_txt=True,
    project="/content/drive/MyDrive/weapon_detection",
    name="test_results"
)
results
```

```python
for i, result in enumerate(results):
    im_with_boxes = result.plot()
```

```python
for r in results:
    boxes = r.boxes
    for box in boxes:
        confidence = float(box.conf)
        class_id = int(box.cls)
        class_name = model.names[class_id]
        print(f"Detected {class_name} with {confidence:.2f} confidence")

    if len(boxes) > 0:
        avg_conf = sum(float(box.conf) for box in boxes) / len(boxes)
        print(f"Average confidence: {avg_conf:.2f}")
```
